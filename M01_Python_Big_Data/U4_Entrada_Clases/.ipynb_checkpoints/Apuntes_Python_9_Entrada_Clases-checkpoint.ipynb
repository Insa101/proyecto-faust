Para poder realizar realizar estas acciones, usaremos el objeto ***File***

> El objeto ***File*** representa una conexion a un fichero determinado y nos permite **leer** su contenido (haciendo uso de ***"r"***) o **escribir** en el (haciendo uso de ***"w"***).
>


A continuacion vamos a abrir un fichero y leerlo, en la siguiente parte explicaremos el detalle del codigo:

```python
import os
DIR_U_UX_Material_extra = os.path.join("..","U_UX_Material_Extra","Python_MaterialAdicional","U09_aux")
fichero_ejemplo= os.path.join (DIR_U_UX_Material_extra, "lorem_ipsum_ext.txt")
abrir = open(fichero_ejemplo, "r")
l_linea_1 = abrir.readline()
print(l_linea_1)
abrir.close()
```

¿Que ha sucedido?

Pues bien esto es lo que ha pasado, *paso a paso*:

- 1º Importamos *os* ya que vamos a necesitar hacer uso de modulos suyos en la *llamada* al fichero
> Usamos el ***import os***
- 2º Hemos configurado una *constante* para denominar una ruta relativa donde se ubica el fichero
> Hacemos uso de la sentencia ***os.path.join()***
> 
>> Es interesante saber que hemos utilizado ***os.path*** que es el modulo que nos permite consultar archivos y la funcion ***.join()*** que nos permite vincular un ruta / archivo a una variable
>> 
> > **Asi se definen las rutas relativas para poder hacer uso de ellas**

- 3º Denominamos **la ruta de acceso del fichero a una variable**, haciendo uso de la misma sentencia que antes (*os.path.join()*), de modo que cuando queramos acceder al archivo, lo tendremos localizado
- 4º Denominamos **a la apertura del archivo a una variable** para poder utilizarlo con otras funciones del sistema.
> Esto lo hemos realizado con la funcion ***open()***
- 5º Decimos al sistema que **lea una linea de la variable que permite leer el fichero**
> Esto lo hemos hecho con la funcion ***readline()***
- 6º Imprimimos la linea que ha leido
- 7º Y mas importante, **cerramos de nuevo el archivo o dara ERROR**
> Esto lo hemos realizado con la funcion ***close()***
>

La clave de todo esto, aparte de generar rutas relativas, ha sido el uso de la funcion ***open()*** y ***close()***

## Entrada de datos: Funcion *open()*

La función ***open()*** nos permite abrir un fichero, que recibe **2 argumentos** y nos devuelve **un objeto *file*** que usaremos para acceder y manipular su contenido.

Funciona del siguiente modo:

- En el 1º argumento se indica **el nombre del fichero que queremos abrir**
  > Si el archivo se ubica en un directorio distinto desde el que intentamos abrirlo, **habra que especificar toda la ruta completa**
- En el 2º argumento se indica ***el modo* en el que se va a abrir el fichero**

Se expresa asi:

> ***"pseudonimo" = open ("nombre del fichero que queremos abrir", "modo en el que queremos abrirlo")***

> Ejemplo: en vez de escribir ***"r"*** escribiras ***"rb"***
