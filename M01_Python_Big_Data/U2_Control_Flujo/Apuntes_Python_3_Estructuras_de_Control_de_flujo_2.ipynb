# Apuntes de Python: Continuacion a las estructuras de control de flujo 2

En programacion se pueden utilizar estructuras de codigo que permiten modificar el flujo de ejecucion de las instrucciones o metodos dados. Lo que nos permite dotar de una estructura que permite que el programa se ejecute en nuestro beneficio segun el resultado buscado.

A este tipo de programacion se le llama **programacion estructurada**.

Se realiza mediantes estructuras de control que controlan el flujo permitiendo que haga una cosa u otra una vez dependiendo:
- De la respuesta de la evaluacion que nos de
- Lo que nos interese como programadores

Hay diferentes tipos:

- **Ejecutar Condicional:**  en base a una condicion, ejecutar un grupo de codigo u otras sentencias
  > *if(...):*
  >
  > *else:*
- **Ejecutar en funcion de una variable:** dependiendo del valor de una variable, ejecutar un grupo de codigo u otras sentencias
  >*select(...):*
  >
  >*case:*
- **Ejecutar mientras**: si se cumple una condicion, ejecutar un grupo de codigo u otras sentencias
  > *while(...):*
- **Ejecutar hasta**: hasta que se cumpla una condicion, ejecutar un grupo de codigo u otras sentencias
  > *while True:*
  >
  > *if(..):*
  > 
  > *break*
- **Ejecutar un numero determinado de veces**: durante un tiempo especifico que se cumpla una condicion, ejecutar un grupo de codigo u otras sentencias
  > *for(...):*

Para que estas estructuras funcionen necesitan disponer de **variables de control** que iran cambiando para que el flujo de ejecucion se ejecute tantas veces como nos interese.

Caracteristicas:

- Todas las estructuras de control disponen de una unica entrada.
- Existen ***secuenciales***, ***condicionales***, ***de seleccion***, ***iterativas*** y ***de control avanzadas***



## Estructuras Secuenciales

Son aquellas que se ejecutan de forma secuencial en tiempo de ejecucion, por tanto, no hay nada que interrumpa la ejecucion de la secuencia.

Una estructura secuenciada puede ser un bloque codigo donde hay variables y funciones:

```python
n = 0
m = 0
v = 0
n = input("dime que es n")
m = input("dime que es m")
v = input("dime que es v")
print ( "pues la n era = a ", n, "la m era = a", m, "la v era = a" ,v)
```

## Estructuras Condicionales

Son aquellas en las que se tiene que dar una evaluacion (condicion) que solo puede ser verdadero o falso [**True** o **False**] (*booleanos*).

Para esa evaluacion utilizaremos los **operadores relacionales** o los **operadores logicos**

### Operadores Relacionales:

Evalua 2 valores y te indica si es *True* o *False*

- "==" es igual que
  > un solo "=" es un simbolo de asignación, no de evaluacion, por eso es neecsario que sea doble
- "!=" distinto que
- "<" menor que
- ">" mayor que
- "<=" menor o igual que
- ">=" mayor o igual que
  > Se escrbe de este modo ya que en otros lenguajes existe el *puntero* que se escribe "=>" y se busca evitar esto.
  
### Operadores Logicos:

Se utilizara para agregar 2 condiciones/comparativas de una misma evaluacion

- "and" hara que tengan que cumplirse las 2 condiciones para ser cierto
- "or" hara que tenga una de las 2 condiciones para ser cierto
- "xor" es una doble negación, y si en ambos casos es cierto, entonces sera falso

```python
if(12>3):
    print("hello")
```

```python
if (20 == 40):
    print ("no")
elif (20 > 50):
    print ( "adios")
elif (20 < 50 and (10 + 3) > 5):
    print ("lo conseguiste wey")
```

## Estructuras de Selección

Son aquellos que ejecutan un bloquer de instrucciones u otro, o saltan a un subprograma o subrutina segun se cumpla o no una condicion.

En este tipo estan incluidas las **estructuras de control condicional**.

```python
n = 10
y = 50
if (n == y):
    print ("no")
elif (20 < 50 and (10 + 3) > 5):
    print ("lo conseguiste wey")
    n = y + 10
    print ( "ademas n = ", n)
elif (n > 50):
    print ( "adios")
```

Pero existen mas esctructuras de control de selección como son las **Select-Case**

Son aquellas que permite ejecutar una accion entre varias en funcion del valor de la expresion.
> Es un alternativa al *if* cuando se cumple la misma expresion con diferentes valores.

Funciona del siguiente modo:
- **Defines una variable:** o *select()* es una selección que se recoge del usuario o es una entrada, que nos va a permitir evaluar y ejecutar un bloque
- **Evaluaciones de la variable:** o *case()* seran los casos en los que se evaluara los valores de *select* y se ejecutara un bloque especifico si coincide el valor
- En cuanto se da la primera coincidencia con alguno de los casos, **se sale de la estructura**
- Si no se encuentra ninguna coincidencia se ejecuta ***case-else***.

> Se puede incluir un *break* para que se salga de la ejecucion en un bloque de sentencias

```python
n = input ( " di algo:")
if (n.isdigit() == True):
    print ("es un numero")
elif (n.isdigit() == False):
    print (" no es un numero")
    n = int(input ("ahora si di un numero"))
    if ( n >= 10):
        print ("es mayor a 10")
    elif ( n <= 2):
        print ( "wow es pequeño")
```

## Estructuras iterativas

Son aquellas en las se realiza una repeticion si se cumple una condicion, iniciando un bloque de instrucciones hasta que deja de serlo o se para.

Estas son ***while (..):*** y el ***while True*** y el ***for (..) in (..)***

Minimo se ejecutan una vez.

 > **VIP: IDENTAR CORRECTAMENTE PARA QUE LOS LOOPS SALGAN BIEN, SINO NO FUNCIONARA CORRECTAMENTE**
>
> En el ejemplo siguiente al no estar bien identado lo que sucede es que los condicionales indicados, no estan dentro del bucle, a no estar dentro del bucle el bucle solo dispone de una entrada input para poder salirse y ya entonces salta a las siguientes lineas de codigo. Dado que la condicion del *while():* y del *if():* son la misma, al salir del bucle la ignora y va directamente a la siguiente condicional.


```python
n = input( " digamelon")
print (" muy bueno!! contesta a lo de abajo")
while n != "galleta":
    n = input ( "que cosita es redonda y marron?")
if ( n != "galleta"):
    print("es otra cosa, venga intentalo de nuevo")
elif (n == "galleta"):
    print ( "bien jugado")
```

> Mientras que si la identas correctamente **haces que las condicionales se tengan en cuenta para cerrar y abrir el loop** haciendo que se transcurra del modo adecuado.

### Bucle *while():*

```python
n = input( " digamelon")
print (" muy bueno!! contesta a lo de abajo")
while n != "galleta":
    n = input ( "que cosita es redonda y marron?")
    if ( n != "galleta"):
        print("es otra cosa, venga intentalo de nuevo")
    elif (n == "galleta"):
        print ( "bien jugado")
```

### Bucle *while True*:

```python
i = True
while i == True:
    j = input ("es otra prueba, di algo")
    if (j != "charmander"):
        print ("error, prueba de nuevo, es un pokemon rojo")
    elif ( j == "charmander"):
        print ( "genial")
        i = False
print ( " por tanto j era igual a " , j)
```

### Bucle *For*:

La logica para este tipo de bucles es la siguiente:

> el bucle se realizara siempre que para (*for*) el valor buscado (*nombre de la variable que va a ser evaluada*) en (*in*) periodo o rango concreto ( *nombre de lista, tupla,...* o *range( numero inicial, numero final)* exista la posibilidad
>
> > Recuerda que **la variable del bucle** es mejor que no este definida o sea 0, para permitir al bucle, cambiarle de valor para que realice iteraciones.
> >
> > Tambien puedes intentar evaluar una variable que ha obtenido un valor y realizar una iteracion hasta que se de el valor exacto o un rango de valores acordes a lo que buscas (como un rango de edad en un proceso de evaluacion de personas sobre si son adultas o no) 

```python
musica = ["canto del loco", "bisbal", "mikel erentxu"]
for i in musica:
    if ( i == "mikel erentxu"):
        print (i,"es un tio cojonudo")
    elif ( i == "bisbal"):
        j = input ("a ver dime algo de este pibe")
        j = "bisbal"
        print (j, "es una puta mierda, espabila")
    else:
        print (i , "son un grupo de locos")
```
