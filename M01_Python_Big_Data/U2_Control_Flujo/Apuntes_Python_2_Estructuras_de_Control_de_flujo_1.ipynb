## Control de flujo 1: Condicionales

Es una estructura que nos permite decidir si se ejecuta o no una o varias operaciones, dependiendo si se cumple la operacion.

En python se utiliza ***if (...):*** para este uso. 

Explicación detallada:
- La funcion reviara un valor booleano que estara entre parentesis () y terminara hasta los dos puntos ":" (sera el *entonces*)
- Los bloques de instrucciones tienen que estar indentados, esto significara que el codigo en esa identacion esta dentro de la condicion.
- Los condicionales solo con el *if(...):* ya funcionaran, por tanto, no es necesario los *else:* o *elif(...):* para poder hacer que se evalue

```python
valor = 100
n_entradas = input( "diga numero de entradas:")
if (int(n_entradas) < valor):
    print("cada entrada vale" , valor, "esta de acuerdo?")
```

Cuando el condicional no se cumple, el codigo identado debajo es ignorado completamente y se sigue el resto del codigo, **el codigo que no se encuentra dentro de la condicional**

```python
valor = 100
n_entradas = input( "diga numero de entradas:")
if (int(n_entradas) < valor):
    print("cada entrada vale" , valor, "esta de acuerdo?")
print( "jeje esto si")
```

En los condicionales para incluir una clausula extra que nos permita evaluar de nuevo la condicion habra que incluir ***else:***
> Lo que nos dice el el *else:* es que para cuando realice la comprobacion de la condicion o condiciones, si no se cumple ninguna condicion en cualquier caso, el resultado/valor va a ser el que le preceda

De modo que para condicion si queremos tener en cuenta **el resultado si se cumple y el resultado si no se cumple** deberemos utilizar siempre ***if(...):*** y ***else:***
> Hay un modo distinto de encontrar un valor que no se cumple y es utilizando ***if not(...):*** que hace la misma función que el ***else:***

```python
valor = 100
n_entradas = input( "diga numero de entradas:")
if (int(n_entradas) < valor):
    print("cada entrada vale" , valor, "esta de acuerdo?")
else:
    print("son demasiadas entradas")
```

Sin embargo, pueden existir escenarios donde haya mas posibilidades de comprobar una condicion que disponga de mas de 2 tipos de resultado (no solo *True* o *False*) por lo que habra que hacer uso de mas condicionales dentro de la condicion, para ello utilizaremos ***elif(...):***

Explicación en detalle:

- Estas condiciones saltaran unicamente cuando la condicion anterior no se cumpla.
- Se pueden concatenar tantas *elif(...):* como se quiera hasta terminar el condicional

```python
valor = 100
n_entradas = input( "diga numero de entradas:")
if (int(n_entradas) < valor):
    print("cada entrada vale" , valor, "esta de acuerdo?")
elif((int((n_entradas)*valor))>1000):
    print("te pasaste wey")
```

>**Los ampliadores de condicionales se utilizaran unicamente cuando nos interese**
>
>Por tanto cuando utilicemos un *elif(...):* no sera necesario terminarlo con un *else:*

## Control de flujo 2: Bucles

Definición: Los bucles nos permiten repetir un bloque de codigo tantas veces como nos indique una condicion.

Cada repeticion realizada por un bucle se llamara: **iteracion**

Existen 2 tipos:

- ***while(...):*** - Se utilizara para realizar una repeticion mientras una condicion se siga cumpliendo.
- ***for(...):*** - Se utilizara para realizar una repetición que sabemos cuantas veces ha de producirse. Muy utilizada con listas [...]

### Bucle *While*

Es un tipo de bucle que se realizara siempre que **se cumpla una condicion**. O lo que es lo mismo, **mientras se de una condicion sucedera algo**.

> La logica sera mientras (***while***) una condición (***(...)***) se cumpla se realizara las iteraciones (**:**).
>
> Por tanto para cada iteración, el programa se preguntara *¿se sigue cumpliendo la condicion?* y si es *si* seguira hasta terminar la iteracion.

Hay que tener en cuenta que este tipo de bucles **necesitan que su condicion cambie para finalizar** o se haran bucles infinitos. Por tanto dentro de las instrucciones **tendra que existir siempre un cambio en las variables de la condicion** para hacer que la iteracion acabe.

> Hay que tener cuidado con os bucles infinitos ya que pueden crashear un ordenador por el uso continuo de recursos del mismo, dependiendo de la complejidad del mismo.
>
> Para salir de un bucle infinito, deberemos presionar *Ctr + C*

```python
numero = 1
while (numero <= 10):
    print( "ahora llevamos" , numero , ",seguimos")
    numero = numero +1
```

> Importate en la concatenación de "texto" con numeros, hay que tener en cuenta que Python va a reconocer el tipo de caracter que es, de modo que si queremos concatenarlo, o bien definimos a los numeros como numeros enteros *int(...)* y los cnacatenamos con + o bien seran cadenas de texto y habra que concatenarlos con , 

Un modo bastante util de utilizarlo es en programas o bloques de codigo que permitan **evaluar una respuesta dada por un usuario** de modo que hagan que si el usuario no responde de manera adecuada a lo que esta concebido el programa, **le vuelva a redirigir al inicio de la iteracion** donde tenga que volver a facilitarnos una respuesta nueva.

El modo mas interesante de realizar esta funcion es utilizar ***while True:*** ya que hara que siempre que se cumpla lo siguiente, volvera al punto de inicio de la iteracion.

> Para parar este tipo de bucles, haremos uso de ***break***, lo veremos mas adelante

### Bucle *For*

Es un tipo de bulce que se utilizara siempre que **sepamos la cantidad de veces que queremos que suceda**. Es un tipo de bucle que lo que hace es realizar unas instrucciones/ repeticiones recorriendo un serie de elementos, con un punto de inicio hasta un punto final.

Se puede entender la logica de esta cadena de texto asi:

> *Para cada valor (***for "i"***) en el codigo (***in***) de inicio a fin (***(valor1, valor2)***) queremos que se ejecute (**:**) lo siguiente*

Lo que sucedera es que la **variable** que indiques despues del *for* sera una **variable auxiliar** que ira cambiando de valor para que el programa pueda ir ejecutando las instrucciones o repeticiones dentro del rango o lista de valores que le indiquemos en *in*.
> Es interesante nombrar a esta variable con un nombre reconocible o util para que sepamos entender bien la variable y para que sirve

El bucle ira pasando de valor en valor en valor hasta que finalice el bucle, de modo que si:

- es una *lista[...]* evaluara o tomara como valor el 1º elemento de la lista, luego el siguiente, luego el siguiente... hasta finalizarla.
- es una *rango(...)* evaluara o tomara como valor el valor inicial, luego el siguiente, luego el siguiente.... hasta el valor final.

Aspectos a tener en cuenta:

- Utilizar variables en las que el usuario pueda determinar el valor, puede ser una opcion mas interesante que utilizar numeros enteros definidos.
- Se utilizan principalmente este tipo de bucles en las *listas, en los diccionarios y en las tuplas*
- Si utilizamos una ***lista [..,..,..]*** para este tipo de bucles se vera asi

```python
lista_ejemplo = ["cosa1","cosa2","cosa3","cosa4"]
for cosa in lista_ejemplo:
    print(cosa)
```

> En caso de que queramos utilizar el bucle con **numeros** utilizaremos la funcion ***range(...,...)*** que genera una secuencia de numeros finita.
> 
> Siempre se incluyen un 2 valores/numeros, el de inicio y el de fin. El de fin nunca lo tendra en cuenta, por tano si queremos que lo cuente deberemos expresarlo con un "+ 1".
>
> La funcion trabaja del siguiente modo, cuando tu le indicas el ***valor*** que tiene que tener en cuenta ***for*** va a situar ese valor en el inicio del bucle, o lo que es lo mismo, lo centrara como ***valor inicial*** y cuando haya ejecutado lo que sigue el bucle, volvera e iniciara el bucle con el ***valor + 1***, asi hasta llegar al ***valor de fin***
>

```python
for valor in range (2,5):
    print(valor)
```

A continuación se iran explicando y ejemplificando algunas caracteristicas de la función ***range(...,...)***
>
> Generalmente ira definida entre **2 numeros**, **uno de arranque y otro de parada**. Atención: El numero de arranque lo tiene en cuenta pero **el de parada no**, por lo que lo excluira. Se escribira asi ***range (... , ...)***.
>

```python
for i in range(1,20):
    print(i)
```

> Se puede utilizar indicando unicamente **un valor**, en ese caso lo tomara **el numero "0" como arranque y el numero dado como el numero de parada**.Se escribira asi ***range (...)***.
>

```python
for i in range(3):
    print(i)
```

> Tambein es posible **incluir otro numero que limite o espacie la secuencia de numeros del rango**, en ese caso se escribira en el ultimo lugar de la secuencia y se escribira asi ***range(... , ... , ...)***. 

```python
for i in range(1,20,5):
    print(i)
```

## Control de flujo 3: Anidación de bucles y condicionales

Se pueden generar concatenaciones de bucles y condicionales anidandose (bloque de codigo dentro de otro bloque de codigo, que se ejecutara si dependiendo del bloque en el que esta incrustado) 

```python
for i in range (2,5):
    for j in range(3,i+2):
        print(j, end = "")
```

Dentro de la anidación, dado que se genera un flujo de instrucciones ordenado entre las distintos bloques de codigo definidos, **es posible romper este orden y alterarlo de modo que se pueda volver a la anterior bucle**.

Para ello utilizaremos ***break***

> La logica para utilizar *break* sera que se utiliza para que cuando se haya cumplido una condicion o se haya dado un caso del modo que necesitamos, **se corte en seco el resto de la iteracion*** . Esto hara que vuelva al ***punto de inicio del bucle (si no se ha cumplido una condicion)*** o bien ***salga del bucle (si se ha cumplido una condicion)***.

En una estructura que usemos el *break* generalmente se usara asi:

- *while(...):* empieza el bucle
  > Aqui podemos utilizar o un *while True:* o una variable que cambie de valor
- Uso de condicionales u otros bucles
- *break* para romper el codigo y volver a iniciar la iteracion

Lo que hace es generar un salto en el orden de las instrucciones, del mismo modo que genera un condicional o un ***elif(...):*** pero de una manera mas elegante y concisa.

### Bucle *while True* , para evaluar respuestas del usuario

Se suele utilizar para la **evaluacion de una respuesta de un usuario** utilizando condicionales y generando un bucle infinito, hasta que el usuario de una respuesta adecuada a lo que buscais para que pueda avanzar, se le saque del bucle con un *break* y asi termina la iteración.

Tiene como componentes:

- *while True:* para iniciar el bucle
- *variable = input(...)* para consultar al usuario por la respuesta que se espera
- *if(...):* para evaluar la respuesta
- *break* para romper el bucle y dejar de hacer iteraciones


```python
while True:
    respuesta = input ( "pregunto:")
    if (respuesta == "hola"):
        print (respuesta)
        break
    elif (respuesta == "mal"):
        print ( "no es lo que esperaba")
        break
    elif (respuesta == "calla"):
        print ("jodeeer")
print ("ya esta")
```

Ahora bien, si queremos anidar diferentes condicionales o bucles, tenemos que tener muy presente como introducrilos y romperlos.

Ya hemos hablado de utilizar *break* para ello, pero tambien podemos hacer **que el cambio de una variable que inicia el bucle sea el que lo rompe**

```python
variable = True
while (variable == True):
    if (variable == 0):
        print ("es un numero")
    else:
        print (" es una palabra")
        variable = False
```

Un modo de hacerlo es utilizando el bucle *while True:* + otro bucle que depende del valor de una variable.

>Haciendolo asi lo que nos permitira es que al suceder el bucle inicial, que dependera del valor de la variable, podemos ir generando iteraciones en sucesivas mediante un bucle infinito con condicionales que una vez cumplidas, podemos cortar + hacer que el valor de la variable del bucle inicial cambie.

Lo hacemos de este modo y sin utilizar un *break* al final del codigo que rompa el bucle inicial, porque sino lo que provocaria es que las iteraciones se dejarian de dar, no se haria el bucle. **Sin embargo al cambiar el valor de la variable** hacemos que las iteraciones paren ya que el bucle no reconoce que se de la condicion inicial, sin necesitar de un *break* para romper ese bucle.

> La logica para esto es que si el trigger inicial ha sido en base a una condicion de un valor de una variable, sera necesario en algun momento de la anidación que ese valor cambie y asi no vuelva a darse la condicion.

```python
v1_bucle = True
while (v1_bucle == True):
    print ( "empeiza el bucle")
    while True:
        respuesta = input ( "pregunto:")
        if (respuesta == "hola"):
            print (respuesta)
            v1_bucle = False
            break
        elif (respuesta == "mal"):
            print ( "no es lo que esperaba")
            v1_bucle = False
            break
        elif (respuesta == "calla"):
            print ("jodeeer")
            v1_bucle = False
            break
print ("ya esta")
```

Por ultimo, se puede realizar un salto mediante un *break* y ademas **mantenerse en el bucle** para que simplemente salte la siguiente iteracion sin salir del bucle completamente.

Para ello utiliaremos ***continue***

Esta sentencia se utiliara para bucles *while(..):* como bucles *for(...)in:*

Lo que hara es que si se cumple la condicion, **se saltara el orden del bucle ignorando los valores indicados en el condicional y continuara con el bucle** con los valores que no cumplan con la condicion.

> la logica es que cuando en un bucle, al comprobar el valor de esa iteracion (*valor1*) si cumple con la condicion y existe un *continue*, en vez de seguir con la iteracion, saltara el valor (*valor1*) y obviara el resto del codigo indentado, de modo que terminara al iteracion y volvera a empezar el bucle con el siguiente valor (*valor2*) para realizar la iteracion.

```python
coches = ["maseratti" , "ferrari", "lambo"]
for coche in coches:
    if (coche == "ferrari" or coche == "maseratti"):
        continue
    print (f"es la polla {coche}")
```

Dado que no siempre vamos a poder hacer este truco, el metodo correcto es hacer uso del metodo de excepciones que se compone de:

- ***try:*** ejecutara las lineas de codigo a modo de evaluador y si las lineas de codigo que contenga, no permiten que se den o se cumpla, lo considerara una excepcion. Evitando que se rompa el programa
  > Como recomendacion, en la funcion es interesante incluir los distintos condicionantes que pueden generar excepciones, para que el programa no se rompa, de modo que si hay varias posibilidades deberia haber al menos mas de un condicional si pueden suceder
- ***except:*** capturara el error hecho por la función *try:*
  > Asi conseguiremos que el codigo no se rompa , y segun como lo configures, no salir del bucle para volver a iniciarlo

```python
cerrar = True
while cerrar == True:
    numero = input ("di un numero entre 5 y 10")
    try:
        verdad = int (numero)
        if (verdad >5 and verdad <10):
            print ("dale")
            cerrar = False
        else:
            print( "espabila")
    except:
        print ("a ver esto")
```
