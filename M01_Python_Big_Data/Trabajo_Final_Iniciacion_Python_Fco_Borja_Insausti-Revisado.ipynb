## Ejercicio 1

### Enunciado ejercicio 1

Crea un programa que solicite al usuario introducir un número entero positivo y lo almacene en una variable.

Si lo introducido por el usuario:

- a) **No es un número** se deberá mostrar por pantalla un mensaje de error que le informe de que lo introducido no es un número, similar a: “El dato introducido no es número. Por favor, vuelva a intentarlo”. Tras ello, el programa deberá volver a solicitarle que introduzca un número entero positivo.
  
- b) **No es un número entero** se deberá mostrar por pantalla un mensaje de error que le informe de que lo introducido no es un número entero, similar a: “No ha introducido un número entero. Por favor, vuelva a intentarlo”. Tras ello, el programa deberá volver a solicitarle que introduzca un número entero positivo.

- c) **No es un número entero positivo** se deberá mostrar por pantalla un mensaje de error que le informe de que lo introducido no es un número entero positivo, similar a: “No ha introducido un número entero positivo. Por favor, vuelva a intentarlo”. Tras ello, el programa deberá volver a solicitarle que introduzca un número entero positivo.

- d) **Es un número entero positivo*** deberás crear una función que lo reciba y compruebe si el número es primo o no. La función deberá devolver un mensaje que informe al usuario del resultado.

> *Recuerda: Un número primo es un número entero divisible tanto por sí mismo como por el 1.*

### Desarrollo Ejercicio 1

```python
# 1º Definimos función para identificar los numeros primos

    # Recibe un número como parámetro
    # Haremos un bucle "for" para comprobar si el número es divisible entre 2 con resto 0. En caso de que solo sea 0 entre 1 y si mismo, sera primo
    # Convierto el resultado en "booleano" para poder hacer uso de él en la función.

def n_es_primo(n):
    if n <= 1:      
        return False
    for i in range (2,n):
        if n % i == 0:
            return False
    return True
    
# 2. Genero la funcion con el programa que consultara al usuario el numero y comprobara si es primo

        # Usaremos un bucle "while" para generar la iteracion con el usuario
        # Pediremos al usuario introducir un numero
        # Utilizaremos el Try/ Except para comprobar que el valor otorgado es un número natural positivo y no cualquier otra opción
        # Se evaluara mediante condicionales "if" si el numero es entero y positivo
    
def introducir_numero():
    print ( f" ---Ejecutando Programa: introducir_numero ---")
    while True:
        numero_usuario = input (f" Por favor, introduzca un numero entero:")
        try:
            comprobar_float = float(numero_usuario) 
            if comprobar_float % 1!= 0:
                print("Error: No ha introducido un número entero. Por favor, vuelva a intentarlo.")
                continue
            elif comprobar_float < 0:
                print("Error: No ha introducido un número entero positivo. Por favor, vuelva a intentarlo.")
                continue
            numero_usuario_ok = int(comprobar_float)
            if n_es_primo (numero_usuario_ok):
                print (f"El número {numero_usuario_ok} es primo")
                print (f" ---Fin del Programa: introducir_numero ---")
            else:
                print (f"El número {numero_usuario_ok} no es primo")
                print (f" ---Fin del Programa: introducir_numero ---")
            break
        except ValueError:
            print("Error: El dato introducido no es número. Por favor, vuelva a intentarlo.")
```

## Ejercicio 2

### Enunciado ejercicio 2



Con los datos mostrados en la tabla anterior crea un programa que:

- Muestre por pantalla cuántos alumnos suspendieron cada asignatura. Ejemplo: Inglés se ha suspendido por 3 alumnos
- Realiza la media de las notas de cada alumno y muéstralas por pantalla. Ejemplo: El alumno1 ha tenido una nota media de 7.25
- Muestra por pantalla los nombres de los alumnos que han aprobado el curso, sabiendo que para aprobar necesitan obtener una nota media igual o superior a 5. Ejemplo: Los alumnos que han aprobado el curso son: Alumno1, …..

### Desarrollo ejercicio 2

```python
# 1. Definimos el diccionario y las variables

    # Definimos el diccionario con los datos fijos del enunciado
    # Definiremos la funcion
    # Creamos contadores a cero para los suspensos de cada asignatura
    # Creamos una lista vacía para guardar los nombres de los que aprueben el curso

notas_alumnos = {"Alumno 1": [ 8, 8, 9, 4],
                 "Alumno 2": [7, 6, 7, 2],
                 "Alumno 3": [10, 7, 8, 9],
                 "Alumno 4": [4, 4, 3, 2],
                 "Alumno 5": [9, 8, 9, 6]
                }
def evaluar_asignaturas (notas_alumnos):
    latin_suspensa = 0
    castellano_suspensa = 0
    frances_suspensa = 0
    ingles_suspensa = 0
    nota_media = 0
    alumnos_aprobados = []
    
# 2º. Generamos el programa que valida las notas

    # Recorremos el diccionario extrayendo clave (alumno) y valor (notas) al mismo tiempo
        # Comprobamos índice 0 (Latín), Comprobamos índice 1 (Castellano), Comprobamos índice 2 (Francés),Comprobamos índice 3 (Inglés)
        # Utilizaremos "items()" para obtener los nombres de los alumnos y sus asignaturas por separado
    # Calculamos la media: suma de las notas / cantidad de notas
    # Imprimimos la media de este alumno por pantalla
    # Comprobamos aprobados: Si la media es aprobada (>= 5), guardamos el nombre en la lista
        # Utilizaremos un ".append()" para anexionar a los alumnos aprobados
    
    print (f" ---Iniciando Programa: Evaluar notas---")
    print (f" ---Primero se mostrarán las medias de los alumnos---")
    for alumno, notas in notas_alumnos.items():
        if notas[0] < 5:
            latin_suspensa = latin_suspensa + 1
        if notas[1] < 5:
            castellano_suspensa = castellano_suspensa + 1
        if notas[2] < 5:
            frances_suspensa = frances_suspensa + 1
        if notas[3] < 5:
            ingles_suspensa = ingles_suspensa + 1
        nota_media  = sum(notas) / len(notas)
        print  (f" El {alumno} tiene una nota media del curso de {nota_media}")
        if nota_media >= 5:
            alumnos_aprobados.append(alumno)

# 3º Mostramos resultados finales (Fuera del bucle)

    # Imprimimos cuántos suspensos hubo en cada asignatura
    # Imprimimos la lista completa de alumnos que han aprobado el curso
    
    print (f" ---Ahora mostraremos las asignaturas suspensas de los alumnos---")
    print(f"Latín: {latin_suspensa} suspenso")
    print(f"Castellano: {castellano_suspensa} suspenso")
    print(f"Francés: {frances_suspensa} suspenso")
    print(f"Inglés: {ingles_suspensa} suspensos")
    print (f" ---Por último indicaremos los alumnos que hayan aprobado el curso---")
    print(f"Los alumnos que han aprobado el curso son: {alumnos_aprobados}")
```

```python
evaluar_asignaturas(notas_alumnos)
```

### Desarrollo del ejercicio 3

```python
# 1º Importamos random, definimos las variables y Definimos la funcion de las tiradas aleatorias

    # Necesitaremos que en cada variable se calcuque una variable numero int aleatoria, por lo que lo primero importaremos la libreria random 
    # Una vez importada habra que configurar una funcion que permita calcular un numero aletorio y guardarlo en una variable
    # Por un lado necesitamos una variable que contendra un numero entero por dado
    # Por otro lado necesitamos un contador de victorias para cada jugador

import random
def tirador_dados ():
    return random.randint (1 , 6)
def juego_dados_aleatorios():
    contador_victorias_j1 = 0
    contador_victorias_j2 = 0

# 2º Definimos el cuerpo del programa para que tire los dados :

    # Sera necesario crear un programa que tenga los siguientes elementos:
        # un bucle "while" para iterar tantas partidas sean neecsarias para que se den los supuestos en los que termina el programa
        # Indicaremos las normas de juego
        # incluiremos los dados y sus "tiradas", mas una variable para "recoja" el ganador de la ronda para utilizarlo luego
            # Haremos uso de la funcion "tirador_dados()" para ello
    
    print ( " ---- Inicializando juego de dados para 2 jugadores -----" )
    print ( " ---- Reglas: ---- " ,  
           "\nSe lanza 4 dados, 2 por cada jugador" ,
           "\n1º Si alguno de los números obtenidos por los jugadores coincide, el Jugador1 ganará la ronda." ,
           "\n2º Por el contrario, si ninguno de los números coincide ganará el Jugador2. " 
           "\nEl primero que gane 3 rondas, gana la partida. " 
          )
    while True: 
        dado_l_j1 = tirador_dados ()
        dado_r_j1 = tirador_dados ()
        dado_l_j2 = tirador_dados ()
        dado_r_j2 = tirador_dados ()
        ganador_ronda = ""

# 3º Desarrollamos los evaluadores de resultados. Seran neecsarios dos evaluadores de condiciones que compruebe:
        # Primero se comprobaranlos resultados de los dados, 
            #Para ello crearemos una variable que recoja todos los supuestos en los que coinciden los dados
            #Despues se comprueba si alguno de los dados coinciden
            # AUmentara el contador de victorias de cada jugador dependiendo del resultado
        # Incluiremos los resultados de cada ronda, el campeon de la misma y como van los marcadores            
        
        dados_coinciden = (dado_l_j1 == dado_r_j1) or (dado_l_j2 == dado_r_j2) or ( dado_l_j1 == dado_l_j2) \
                            or (dado_r_j1 == dado_l_j2) or (dado_r_j1 == dado_r_j2) or (dado_l_j1 == dado_r_j2)
        if dados_coinciden:
            contador_victorias_j1 = contador_victorias_j1 + 1
            ganador_ronda = "Jugador 1"
        else:
            contador_victorias_j2 = contador_victorias_j2 + 1
            ganador_ronda = "Jugador 2"
        print( "-" * 40)
        print ( f"\nResultados:el jugador 1 ha obtenido estos dados [{dado_l_j1}, {dado_r_j1}]" ,
               f" el Jugador 2 ha obenido estos datdos [{dado_l_j2}, {dado_r_j2}]"
              )
        print ( f"Gana la ronda: {ganador_ronda}")
        print( f"Marcador Global -> Jugador 1 lleva {contador_victorias_j1} victorias | Jugador 2 lleva {contador_victorias_j2} victorias")
        if contador_victorias_j1 == 3 or contador_victorias_j2 == 3:
            print ( "Enhorabuena has ganado la partida jugador 1")
            print( "-" * 40)
            print ( "Ahora se le consultara si quiere iniciar otra partida")

# 4º Generamos un consultor final para repetir partida

        # Aqui incluimos otro bucle while True para validar la pregunta. 
        # Dado que no vamos a cambiar el tipo de dato, podemos hacer una validacion por condicional "if"
            # Pediremos al usuario que responda SI o NO. Modificaremos su respuesta para estandarizarla y evitar errores.
            # Cambiaremos el formato de la respuesta para que nunca de error en caso de que lo escriba de otro modo
            # Si contesta SI, se inicia la partida. Si contesta NO se acaba. Si dice otra cosa, saltara el error.
            
            while True: 
                repeticion_partida = input (" ----- ¿Desea volver a iniciar una partida? Responda con un SI o NO-----" ).upper().replace("Í","I")
                if repeticion_partida == "SI":
                    contador_victorias_j1 = 0
                    contador_victorias_j2 = 0
                    print ( " ---- Inicializando nueva partida ----" )
                    break
                elif repeticion_partida == "NO":
                    print ( " ---- Muchas gracias por jugar con nosotros ----" )
                    return
                else:
                    print ( " Disculpa me temo que la respuesta introducida no es valida, se te volvera a consultar")
```

## Ejercicio 4

### Enunciado Ejercicio 4

Crea una clase llamada PERSONA con las siguientes propiedades: nombre, apellido, nacionalidad y una función que se llame saludar que muestre un saludo
por pantalla.

Después crearás una clase llamada ITALIANO, subclase de PERSONA, que tendrá como propiedad: idioma principal cuyo valor estará ya definido y será italiano. Además, la clase ITALIANO establecerá que el valor de la nacionalidad (propiedad que hereda de la clase PERSONA) estará ya definido y será siempre: italiana. 

Por último, la clase incorporará un método llamado saludar que mostrará por pantalla un saludo similar a: “Bonjorno! Mi nombre es: ……... ……… , soy de nacionalidad italiana y mi idioma principal es italiano.”

Tras la creación de las dos clases, deberás crear un programa que:
- Solicite al usuario los datos necesarios para crear un objeto de la clase Italiano. Recuerda que el idioma y la nacionalidad ya tienen valores definidos por lo que no se le deberían preguntar al usuario.
- El programa deberá contener un método que reciba los datos introducidos por el usuario para crear con ellos un objeto/instancia de la clase italiano.
- Por último, se deberá mostrar por pantalla el mensaje del saludo, llamando a la función saludar desde el objeto de la clase ITALIANO que se ha creado.

### Desarrollo Ejercicio 4

```python
# 1º Debemos crear la class padre de la que heredaran metodos y atributos luego la class hija. 
        # Crearemos un metodo constructor para sus 3 atributos
        # crearemos una funcion que sea saludo

class Persona:
    def __init__ (self, nombre, apellido, nacionalidad):
        self.nombre = nombre
        self.apellido = apellido
        self.nacionalidad = nacionalidad
    def saludar (self):
        print ( f" Buenos dias mundo, mi nombre es {self.nombre} {self.apellido}")

# 2º Crearemos la class hija llamada Italiano, que heredara algunos atributos y metodos y otros seran definidos. 

    # el metodo constructor heredara los atributos nombre y apellido de la clase Persona
        # La nacionalidad sera definida como "Italiana" en el metodo constructor
        # el idioma "Italiano" estara predefinido
    # definiremos un nuevo saludo en italiano, que incluya la nacionalidad y el idioma.

class Italiano (Persona):
    def __init__ (self, nombre, apellido):
        super().__init__(nombre, apellido, "italiana")
        self.idioma = "italiano"
    def saludar_italiano (self):
        print ( f"Bonjorno!, mi nombre es {self.nombre} {self.apellido}.",
               f"Soy de nacionalidad {self.nacionalidad} y mi idioma principal es {self.idioma}"
              )

#3º Creamos un programa que permita al usuario crear un objeto de la clase Italiano. 

    # Utilizaremos input() para poder solicitar los atributos nombre y apellido
    # Utilizaremos las variables para definir los atributos de Italiano
    # Invocaremos al saludar.italiano
    
def crear_italiano ():
    print ( "---- Iniciando programa para definir persona Italiana ----")
    nombre = input ( "Indique su nombre de pila o principal:")
    apellido = input ("Indique ahora un apellido, su primer apellido")
    persona_italiana = Italiano (nombre, apellido)
    persona_italiana.saludar_italiano()
```

```python
crear_italiano()
```

### Desarrollo del Ejercicio 5

```python
# 1º Haremos una función que genere un tupla con numeros aleatorios

    # Utilizaremos la libreria "random" para ello y un bucle "for" para iterar 15 veces
    # Tambien dado que las tuplas son inalterables, primero haremos una lista y luego la convertiremos a tupla
    # Devolveremos la tupla para utilizarla en la función siguiente

import random
def numeros_ganadores():
    lista_numeros = []
    for n in range (15):
        ganador = random.randint (0, 100)
        lista_numeros.append(ganador)
    tupla_ganadores = tuple (lista_numeros)
    return tupla_ganadores

# 2º Generamos el programa donde le solicitaremos al usuario que introduzca un numero entero

    # La tupla ha de estar fuera del bucle para que sea una lista inalterable en cada intento
    # Lo primero sera incluir la validacion de respuesta para que funcione el programa, al solicitar un numero, mediante condicionales y bucle while
    # Hay que hacer otro bucle "while" para la validacion de respuesta.
        # Utilizaremos el Try/ Except para comprobar que el valor otorgado es un número natural positivo y no cualquier otra opción
        # Se evaluara mediante condicionales "if" si el numero es entero y positivo

def juego_loteria():
    tupla_ganadores = numeros_ganadores()
    while True:
        while True:
            numero_usuario = input (f" Por favor, introduzca un numero entero:")
            try:
                comprobar_float = float(numero_usuario) 
                if comprobar_float % 1!= 0:
                    print("Error: No ha introducido un número entero. Por favor, vuelva a intentarlo.")
                    continue
                elif comprobar_float < 0:
                    print("Error: No ha introducido un número entero positivo. Por favor, vuelva a intentarlo.")
                    continue
                numero_usuario_ok = int(comprobar_float)
                break
            except ValueError:
                print("Error: El dato introducido no es número. Por favor, vuelva a intentarlo.")

# 3º Se mostrara la tupla de numeros ganadores, y el numero mas pequeño y el mas grande

        
        print ( f"-------- Los numeros premiados aparecen a continuacion: --------")
        print ( f"El numero mas pequeño es { min(tupla_ganadores)}" 
                f"\nEl numero mas grande es { max(tupla_ganadores)}"
                f"\nTodos los numeros premiados son los siguientes:",
                f"{tupla_ganadores}"
                )
        
# 4º Ahora habra que comprobar si el numero del usuario esta premiado o no y en caso de que si, de cuanto es el premio.

    # Para ello realizaremos una validacion, calcularemos el premio y mostraremos el resultado por pantalla
    # El premio sera segun indican en el enunciado: 15€ si sale 1 vez, 5€ extra por cada vez de mas que salga el numero
    # En caso de no estar premiado, se le consultara de nuevo al usuario, mediante un bucle "while". 
        #Si dice que si, se vuelve a lanzar el programa.
        #Si dice que no, se cerrara el programa.
        
        if numero_usuario_ok in tupla_ganadores:
            numero_repetido = tupla_ganadores.count(numero_usuario_ok)
            if numero_repetido == 1:
                premio = 15
                print (f"¡Felicidades! Su número: {numero_usuario} se encuentra dentro de la lista de ganadores. Ha ganado un total de {premio}€")
                print( "-" * 40)
                return
            elif numero_repetido > 1:
                premio = 15 + (5*(numero_repetido-1))
                print (f"¡Felicidades! Su número: {numero_usuario} se encuentra dentro de la lista de ganadores", 
                        f"y ademas se ha repetido {numero_repetido} veces.", 
                        f" Ha ganado un total de {premio}€"
                        )
                print( "-" * 40)
                return
        else:
            print( "Vaya, parece que tu numero no aparece entre los ganadores, ¿Dispones de otro numero con el que jugar?")
            while True:
                reintento = input (" Indique SI o NO").strip().upper().replace("Í","I")
                if reintento == "SI":
                    print ( " ---- Reiniciando el programa ----" )
                    break
                elif reintento == "NO":
                    print ( " ---- Muchas gracias por jugar con nosotros ----" )
                    return
                else:
                    print ( " Disculpa me temo que la respuesta introducida no es valida, se te volvera a consultar")
```
