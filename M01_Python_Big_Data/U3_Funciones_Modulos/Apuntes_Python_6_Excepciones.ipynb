# Apuntes de Python: Tratamiento de excepciones en Python

En general al programar, debido a la naturaleza de la actividad que trabaja con diferentes datos obtenidos de diferentes formas, se pueden generar errores o intereferencias generadas de la ejecucion de las instrucciones que hayamos generado.

Vamos a matizar antes que nada cual es la diferencia entre ***error vs excepción***:

- **Un error** es todo suceso en el que el programa se comporta correctamente, el comportamiento del usuario es acorde a nuestras expectativas y el programa no esta ejecutandose.

- **Una excepcion** todo aquello que **sucede bloqueando el programa que depende de nuestra falta de prevision**, ya que el programa esta haciendo lo que tiene que hacer, pero no hemos contemplado que el usuario actue de un modo distinto al previsto.



```python
variable = int(input("diga un valor:"))
```

Para poder manejar estas excepciones, podemos hacer uso de **las herramientas de gestion de excepciones** que nos permitiran hacer que el programa se siga ejecutando otro bloque de codigo, teniendo en cuenta esos valores que bloquearian el correcto funcionamiento, pudiendo ser alguna excepcion en particular, varias excepciones en particular o todas las excepciones en general.

El 1º metodo para evitar que surja un error o una excepción si no se puede realizar el codigo que estamos intentando hacer es hacer uso de ***int(..)*** para los numeros enteros o ***float()*** para los numeros con decimales, permitiendo que el sistema interprete que son numeros los valores dados en una función ***input()***

> ***int(input("texto que aparecera al usuario"))***
>
> ***float(input("texto que aparecera al usuario"))***

Habra que tener en cuenta lo siguiente:

> el valor dado por el usuario sera considerado como **una cadena de texto** y habra que indicarle que es un numero ,sino, dara un error o **una excepcion**
>
> En ambos casos el sistema nos los va a entender que es un numero

```python
variable = int(input("di un numero"))
```

Dado que no siempre vamos a poder hacer este truco, el metodo correcto, **ya que el usuario sera imprevisible** es hacer uso del metodo de excepciones que se compone de:

- ***try:*** ejecutara las lineas de codigo a modo de evaluador y si las lineas de codigo que contenga, no permiten que se den o se cumpla, lo considerara una excepcion. Evitando que se rompa el programa
  > Como recomendacion, en la funcion es interesante incluir los distintos condicionantes que pueden generar excepciones, para que el programa no se rompa, de modo que si hay varias posibilidades deberia haber al menos mas de un condicional si pueden suceder
- ***except:*** capturara el error hecho por la función *try:*
  > Asi conseguiremos que el codigo no se rompa , y segun como lo configures, no salir del bucle para volver a iniciarlo

La logica detras de esta expresion es:

- *Funcion a probar* sera la funcion que puede se va a evaluar y puede generar errores que rompa el codigo, de la cual habra que generar excepciones.

- *try:* recibe la orden de *intentar* realizar el siguiente codigo y si por cualquier motivo surge alguna excepcion, permite avanzar.

- *except:* da la orden de ejecutar un codigo distinto, sin romper el codigo anterior ni el bloque donde esta incluido

```python
numero = input ("di un numero entre 5 y 10")
try:
    verdad = int (numero)
    if (verdad >5 and verdad <10):
        print ("dale")
        cerrar = False
    else:
        print( "espabila")
except:
    print ("a ver esto")
```

> **Es muy importante usar el tratamiento de excepciones para el desarrollo de muchas lineas de codigo y asi poder averiguar posibles errores inadvertidos**

Es un modo correcto de actuar porque nos permite varias cosas:

- Nos permite contemplar **diferentes tipos de excepciones** y actuar con cada una de ellas de un modo concreto.
- Nos permite no solo evaluar el valor dado sino, que **tambien nos permite hacer uso del valor**
- Nos permite **reducir lineas de codigo**

Es muy utilizado en bucles ***while == True:*** ya que si necesitamos que el usuario nos facilite un tipo de dato concreto, nos permitira iterar el proceso de solicitud del dato las suficientes veces para que el codigo se ejecute del modo que necesitamos

> ***while == True:***
>
> ***input ("solicitud usuario")***
>
> ***try:***
>
> ***except:***

```python
cerrar = True
while cerrar == True:
    numero = input ("di un numero entre 5 y 10")
    try:
        verdad = int (numero)
        if (verdad >5 and verdad <10):
            print ("dale")
            cerrar = False
        else:
            print( "espabila")
    except:
        print ("a ver esto")
```

Tambien es habitual en este tipo de estructuras con **bucles infinitos**, hacer uso de la sentencia ***break*** de modo que si el dato introducido es casi correcto, permite saltar el resto del metodo para **optimizar el codigo**

```python
while True:
    numero = input ("di un numero entre 5 y 10")
    try:
        verdad = int (numero)
        if (verdad >5 and verdad <10):
            print ("dale")
            break
        elif verdad == 0:
            print( "noo es 0, pon un numero entero entre 5 y 10")
    except:
        print ("a ver esto son palabras, pon un numero")
```

Un uso basico para este metodo de evaluacion es integrar el bloque de codigo en **una funcion personalizada** para poder uso de ella cuando la consideremos necesaria, cada vez que nos pueda interesar, asi conseguimos que el codigo **este optimizado y ahorrar lineas de codigo**

Con este uso, tendremos que introducir la sentencia ***return "nombre variable"*** para que nos permita **mantener el dato introducido y validado**

> Sustituye al ***break*** para capturar el valor y salir del bucle

Luego se incluyen variables que se definirian igual que la funcion, teniendo asi los resultados capturados y podremos hacer uso de ellos.

```python
def tratamiento_excep ():
    while True:
        numero = input ("di un numero entre 5 y 10")
        try:
            verdad = int (numero)
            if (verdad >5 and verdad <10):
                print ("dale")
                return verdad
            elif verdad == 0:
                print( "noo es 0, pon un numero entero entre 5 y 10")
        except:
            print ("a ver esto son palabras, pon un numero")

numero1 = tratamiento_excep()
numero2 = tratamiento_excep()

print (numero1)
print (numero2)
```

Para mejorar el uso de este tipo de metodos en funciones podremos **incluir parametros** que ayuden al usuario a la definicion de valores que tiene que dar.

Para ello tendremos que definir en la funcion **una variable** que es la que se definira cada vez que lo ejecutemos.


```python
def tratamiento_excep (variable):
    while True:
        numero = input (f"{variable} di un numero entre 5 y 10")
        try:
            verdad = int (numero)
            if (verdad >5 and verdad <10):
                print ("dale")
                return verdad
            elif verdad == 0:
                print( "noo es 0, pon un numero entero entre 5 y 10")
        except:
            print ("a ver esto son palabras, pon un numero")

numero3 = tratamiento_excep("puedes decir un numero no? venga")
```
