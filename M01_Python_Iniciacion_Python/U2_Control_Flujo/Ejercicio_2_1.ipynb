### 15. Crea un programa que cuente los números pares e impares de una tupla de números del 0 al 10 y muestra el resultado por pantalla. 

```python
tupla1 = (1,2,4,7,8,4,7,6)
pares = 0
impares = 0
for i in tupla1:
    if i %2 ==0:
        pares +=1
    else:
        impares +=1
print (f"los numeros pares de mi tupla compuesta por estos numeros {tupla1} son en total" , pares, end = "  numeros")
print (f"\n los numeros impares de mi tupla compuesta por estos numeros {tupla1} son en total" , impares, end = "  numeros")
```

### 19. Crea un diccionario llamado menú dónde guardes el nombre de un alimento y su precio. Después crea un bucle que permita solicitar al usuario un alimento y le informe del total del pedido hasta que el usuario ya no quiera pedir más y entonces se le informe del total de la cuenta. El usuario saldrá del bucle si no introduce ningún dato. Si introduce un alimento que se encuentre fuera del menú se le informará que no disponemos de dicho alimento en el menú. 

```python
menu = {"carne":30, "pescado": 45, "lechuga": 4, "agua" : 0 , "vino": 14, "pasta" : 9}
cuenta = 0
verificador = True
while verificador is True:
    alimento_usuario = input ( "por favor solicite algun alimento: ")
    if alimento_usuario in menu:
        cuenta += menu.get(alimento_usuario)
        print (f" la cuenta asciende a un total de ", cuenta)
    elif alimento_usuario not in menu and alimento_usuario != "":
        print ( "lo lamento, ese alimento no se encuentra en nuestro menu, vuelva a intentarlo")
    elif alimento_usuario == "":
        print ("se cerro la sesion, adios")
        verificador = False
```
