## Variables y constantes

Respecto a las variables, python es un programa de tipado dinamico, lo que le permite averiguar que tipo de dato define la variable, en funcion del valor que le hayamos asignado, lo que nos permite ir cambiando el tipo de dato de la variable y que el programa lo pueda ir reconociendo.

Para asignar un valor a una variable se hace incluyendo un signo "=" 
Python permite el mismo valor a distintas variables simultaneamente y distintos valores a distintas variables

Ejemplos

> Podemos definir a las variables un nombre que utilizamos como referencia para indicar un objeto que tenemos en la memoria.

```python
#1 valor 1 variable
nombre = "jose luis"

#1 valor varias variables
marcador = goles = 1

#1 valor para cada variable
partidos_perdidos, partidos_ganados = 0, 3
```

Se pueden intercambiar valores de variables de manera sencilla, unicamente haciendo uso de "=" y las comas ","

Ejemplo

```python
x = 20
y = 15
print("x =", x, "y =", y)

#cambiamos los valores asi

x, y = y, x
print("x =", x, "y =", y)
```

En python **no se pueden definir constantes como tal ya que siempre se puede cambiar el valor**, pero por convencion se decidio que las constantes se denominarian con **MAYUSCULAS**

```python
VERSION = "1.0.0"
```

## Números

Se puede trabajar con distintos numeros:
- *Int* = enteros o numeros reales
- *float* = flotantes o numeros con decimales

Se puede operar normal con ellos

```python
5+8
```

```python
15-20
```

```python
150*82
```

Si la **operacion contiene decimales** , siempre se dara el **resultado como un numeros *flotantes***. La division siempre devuelve numeros flotantes. Si mezclamos varias operaciones entre enteros y decimales, python adaptara el reusltado a lo mas comun.

```python
65/12
```

```python
3.5*45
```

Si queremos una **division entera** se hara con "//"

```python
154//24
```

Un **modulo** o un resto de una división se hara con "%"

```python
25/2
```

```python
25%2
```

Una **potencia** se hara con "**"

```python
25**2
```

Un **valor absoluto** se hara con "abs()"

```python
abs(-15.2358)
```

**Redondear** se hara con "round()". Se incluyen mas cifras decimales poniendo "," en ().

```python
round(58/25)
```

```python
round(58/25,5)
```

## Cadena de caracteres

Python permite poner cadenas de texto de varias maneras

> Se recomienda que para evitar errores en Python se utilice unicamente un estilo de de entrecomillado, o "" o ''.

```python
# Con comillas simples
texto = 'texto'
```

```python
# Con comillas dobles
texto = "texto"
```

```python
# Incluir comillas simples en un entrecomillado doble
texto = "esto es 'texto' dentro de texto"
```

```python
# Incluir comillas dobles en un entrecomillado simple
texto = 'esto tambien es "texto" dentro de texto'
```

Se pueden **concatenar 2 cadenas de texto usando el +**

```python
texto = "1º texto" + ' 2º no texto'
texto + "  3º eso"
```

Tambien se puede repetir una cadena de texto **x veces indicando "*"**

```python
texto = "tex"*3
print (texto)
```

> Si se incluye un operador numerico como + , no hace falta utilizar la funcion print() para que lo muestre en pantalla

Se pueden **contar los caracteres** de una cadena de texto haciendo uso de los corchetes "[]". 
Python empezara a contar desde 0, de modo que el 1º caracter estara en el 0 y no en el 1. Por tanto si estas buscando la posicion 5, en verdad te devolvera la 6.

```python
texto = "esto es texto"
texto[5]
```

Se puede **contar desde un caracter hasta otro**, *una rebanada* como se dice, poniendo : entre el corchete [].

Se excluyen deliberadamente **el primer caracter si el numero es positivo** y **el ultimo caracter si el numero es negativo**

```python
texto = "esto es texto"
texto[2:8]
```

```python
texto = "esto es texto"
texto[0:8]
```

```python
texto = "esto es texto"
texto[2:]
```

```python
texto = "esto es texto"
texto[:8]
```

```python
texto = "esto es texto"
texto[:-2]
```

```python
texto = "esto es texto"
texto[-4:]
```

Se pueden juntar 2 rebanadas del siguiente modo para no perder caracteres

```python
texto = "esto es texto"
texto[:2] + texto [2:]
```

Las cadenas de texto de Python permiten manipular mediante operaciones la cadena ofreciendo una version modificada, sin manipular la cadena original.

> Cambio de cadenas de texto a **MAYUS** sera con ***.upper()***, indicandose delante de la cadena de texto que queramos modificar.

```python
texto = "esto es texto"
texto.upper()
```

> Cambio de cadenas de texto a **minus** sera con ***.lower()***, indicandose delante de la cadena de texto que queramos modificar.

Si quisieras modificar la variable original con la version modifica, deberias indicarselo.

```python
texto = "esto es texto"
texto.upper()
texto = texto.upper()
print(texto)
```

> Si quieres **contar cuantas veces aparece un caracter en una cadena** tendras que utilizar ***.count('...')***

```python
texto = "esto es texto"
texto.count('e')
```

> Si quieres **reemplazar un caracter en una cadena** tendras que utilizar ***.replace('...' , '...')***

```python
texto = "esto es texto"
texto.replace('e', 'x')
```

> Si quieres **partir una cadena de texto mediante un delimitador** tendras que utilizar ***.split('...')***
> 
> Si no damos ningun delimitador por defecto lo separa teniendo en cuenta ***los espacios***
>
> Se puede utilizar la asignacion multiple para recoger cada valor y darselo a una variable

```python
texto = "esto es texto"
texto.split(' ')
```

```python
texto = "esto es texto"
texto.split ()
primera, segunda, tercera = texto.split ()
print(primera)
print(segunda)
print(tercera)
```

> Si quieres **unir mediante una cadena de texto una lista de elementos** tendras que utilizar ***"...".join([...])***

```python
texto = "esto es texto"
texto.split ()
primera, segunda, tercera = texto.split ()
"-".join([primera, segunda, tercera])
```

> Si quieres **comprobar si la cadenea de texto esta compuesta por letras** tendras que utilizar ***"...".isalpha()***

```python
"texto".isalpha()
```

> Si quieres **comprobar si la cadenea de texto esta compuesta por numeros** tendras que utilizar ***"...".isdigit()***

```python
"texto".isdigit()
```

> Si quieres **insertar cadenas de texto en otras cadenas de texto** tendras que utilizar ***"{..}.format(..)***
>
> Se pueden añadir mas de uno añadiendo "," en ()
>
> Puedes hacer referencia a variables o poner un valor

```python
texto = "esto es texto"
texto.split ()
primera, segunda, tercera = texto.split ()
"{} no es un texto nuevo es solo {} y si tienes dudas sera solo {}".format(primera, segunda, 14)
```

> Hay un modo abreviado de uso dentro de una condicion o de una funcion que es indicando ***f"..{}.."***

```python
texto = "quien"
print (f"has dicho que {texto} es quien?")
```

## Valores Booleanos

Los valores booleanos sirven para realizar comprobaciones TRUE o FALSE, en python se hacen asi

```python
esto_es_texto = True
texto_es_numero = False
```

el operador logico ***Y*** se utiliza asi

```python
esto_es_texto = True
texto_es_numero = False
esto_es_texto and texto_es_numero
```

el operador logico ***O*** se utiliza asi

```python
esto_es_texto = True
texto_es_numero = False
esto_es_texto or texto_es_numero
```

los operadores de comparacion seran:
- '<' menor que
- '>' mayor que
- '>=' mayor o igual que
- '<=' menor o igual que
- '==' igualdad, exactitud
- '!=' distintos, no iguales

```python
4<8
```

```python
57>80
```

```python
27<=89
```

```python
7==5
```

## None

Existe un valor en python para definir que una variable no esta definida y este valor es ***None***. 

Se puede comprobar si un variable esta definida mediante el uso de ***is none***

Podemos indicarlo para que el sistema sepa que no esta definido.

Es un tipo de dato que vamos a poder utilizar, hace que sea distinto al *null*

> Podremos utilizar este valor *None* en casos en los que por ejemplo dispongamos de muchos datos con los que trabajar y queramos pasarlos a una tabla numerica, nos permitira trabajar con datos que nos den error o esten vacios.

```python
texto = None
texto is None
```

```python
texto = ""
texto is None
```
