# Apuntes Python: Entradas y Salidas estandar

Generalmente estos metodos son metodos simples de entrada y salida de datos en un programa, comun a todos los programas independientemente del lenguaje.

Se genera un canal de comunicacion donde entra informacion en el programa y antes de cerrarlo, sca esa informacion, procesada o en error. Aqui hay una imagen que define esta conexion:



La entrada de datos o ***stdin*** es una conexin solo de lectura a travs de la que un programa puede recibir y leer datos generados desde el entorno en el que se lanz.

En python se expresa asi:

> ***sys.stdin***
>
> > Ejemplo, cuando lanzamos un programa desde el terminal o consola de texto, es la conexin que recibe la entrada generada desde el teclado.

La salida estndar ***stdout*** es la conexin solo de escritura en la que se escribe por defecto cualquier mensaje o resultado de salida, cuando no se utiliza explcitamente otro fichero o conexin.

> ***sys.stout***
>
> > Ejemplo, cuando hacemos un  ***print()*** se realiza un stout


Por ultimo, existe una salida de error estndar ***stderr***. Es una conexin similar a *stdout*, solo de escritura, y que se utiliza como el canal por defecto en el que se escriben los mensajes de error, excepciones, etc.

> ***sys.stderr***
>
> > Ejemplo, cuando hacemos se ejecuta una excepcion o error
