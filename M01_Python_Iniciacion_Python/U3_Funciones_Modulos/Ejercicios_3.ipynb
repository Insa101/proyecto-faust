```python
def comprobar_menor_10(lista):
    listado_menores_10 = []
    listado_mayores_10 = []
    repetidos =[]
    for n in lista:        
        if n < 10:
            if n not in listado_menores_10:
                listado_menores_10.append (n)
            else:
                repetidos.append(n)
        else:
           listado_mayores_10.append (n)
    return listado_menores_10 , listado_mayores_10 , repetidos
```

> Las horas extras trabajadas se pagan 10€ más que la hora normal.

- Solicite al usuario la cantidad de horas que ha trabajado esa semana y el precio al que le pagan la hora normal.
- Ejecute la función pasándole los valores necesarios como parámetros.

```python
def calculadora_salario_semanal ():
    print ( f" Bienvenid@, esta es una calculadora para calcular su salario por hora bruto de la semana. Empecemos")
    horas = int(input(f" digame, ¿cuantas horas ha trabajado esta semana?"))
    coste = int(input(f" ¿cuanto gana a la hora?"))
    print(f" Indica que esta semana ha trabajado {horas} horas y cobra la hora a {coste} €/hora. Calculando...")
    print ( f" recuerde que su jornada se ha establecido que sea de 35 horas y cualquier extra de esas se cobrara, 10€ adicionales")
    if horas > 35:
        extras = horas -35
    salario_base = (horas-extras) *coste
    if extras >0:
        salario_extra = extras*(coste+10)
    salario_semanal = salario_base + salario_extra
    return print (f" su salario de esta semana ha sido calculado: su salario base es de {salario_base} € y su salario extra es de {salario_extra} € por tanto su salario semanal total es de {salario_semanal} €")
```
