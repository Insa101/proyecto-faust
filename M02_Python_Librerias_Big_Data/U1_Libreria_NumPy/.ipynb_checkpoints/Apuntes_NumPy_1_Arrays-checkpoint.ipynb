## Arrays: Â¿Que son?

Los ***Arrays*** son el tipo de dato basico con el que se trabaja en esta libreria.

> Son objetos que representan vectores o matrices **multidimensionales** en las que todos los elementos **son de un tipo**, nÃºmerico generalmente.
>

Estos objetos son denominados ***ndarray***

Los *Arrays* nos van a dar la posibilidad de realizar operaciones de manera muy sencilla con estructuras de datos mas complejos (*scalars*).

> En Python *basico* existen las listas que nos permite minimamente operar pero en cuanto requiere trabajar funciones en Ã¡lgebra avanzado, no estan diseÃ±adas para poder comportarse como necesitamos, de ahi que nazcan los *Arrays*
>


### Diferencias entre *Arrays* y Listas en Python

Antes de continuar, es importante hacer esta diferenciacion entre estos 2 tipos de estructuras de datos.

- La 1Âº diferencia es que las listas pueden estar compuestas por **datos heterogeneos** y los *Arrays* deben ser **datos homogeneos** (del mismo tipo, *scalar*)

> *lista = ["carlos", 5.4, -6, "P"]*
>
> *ndarray = ([5, 7, 89])* / *ndarray = (["carlos", "p"])* / *ndarray = ([5.4, 7.7, 89.0])*
>
> > Esto es asi para mejorar el rendimiento del uso de este tipo de dato

- La 2Âº es la capacidad de realizar **operaciones matematicas mas eficientes**, debido a varios motivos:

> - Cuando realizas un calculo en una lista, el programa **ha de recorrerla** para realizar el calculo, pero en los *arrays** **no la recorre** solo se realiza como una unidad.
>
> - el *array* consume menos memoria e incluye referencias al tipo de dato
>
> > Es debibo, por asi decirlo ,a que *Python Basico* almacena en su memoria "el valor del dato" + "el tipo de dato" **para cada dato** mientras que *NumPy* almacena "valor del dato" + "un tipo de dato" **para todos los datos del *Array***
> >

- La 3Âº es que se pueden **crear y manipular *Arrays* de diferentes formas**

> Pueden venir desde una lista, podemos generar que dispongan de varias dimensiones o de 1 unica dimension, introducir tuplas...

### Funciones para usar *Arrays*

Existen funciones especificas que podemos utilizar con estas estructuras que nos van a dar la posibilidad de sacarle el potencial que hemos mencionado.

Algunas son:

- ***np.ones()***: Genera un Array con un tamaÃ±o y tipo indicados, con valores 1.
- ***np.zeros()***: Genera un Array con un tamaÃ±o y tipo indicados, con valores 0.
- ***np.random.random()***: Genera un Array, con valores aleatorios entre 0.0 y 1.0.
- ***arange()***: Genera un Array de 1 dimension (vector) de un intervalo dado
- ***reshape()***: Modifica un Array existente.
- ***flatten()***: Copia un Array a un Array de 1 dimension

#### Funcion *np.ones()* (incluido *np.zeros()* y *np.random.random*)

Genera un array de *inicializacion a unos* que genera un array con valores 1.

> En el caso de *np.zeros()* es con valores 0
>
> En el caso de *np.random.random* es con valores pseudoaleatorios entre 0 y 1.

Este(os) Arrays(s) se compone(n) de 2 elementos que los definen: 

- 1Âº **el tamaÃ±o** o *shape* nos indica la forma del array, si dispone de 1 dimension o de mas dimensiones (utilizando las tuplas).
  > Si indicamos un unico valor lo considerara **un vector** con 1 dimension
  >
  > Si indicamos una tupla, lo considerara **una matriz**, cada valor en la misma indicara, **numero de filas** y **numero de valores en la fila** (o *scalars*)
  >
  > > ejemplo: (2,4) = 2 Filas, 4 Valores por fila

- 2Âº **el tipo** o *dtype* nos indica el tipo de dato del array, por defecto sera *numpy.float64*

#### Funcion *arange()*

Generara un Array de 1 unica dimension y un rango de numeros *(int o float).*

Caracteristicas:

- *start* sera el inicio del intervalo (por defecto 0)
- *stop* final del intervalo (*int* o *float*)
- *step* indica la diferencia entre 2 valores del intervalo (por defecto 1)


#### Funcion *reshape()*

Modifica un Array en otro tipo de Array

> Puedes generar Arrays de 1 dimension (vectores) a raiz de Arrays de mas dimensiones (matrices), viceversa,...

Caracteristicas:

- *a* indica el Array a modificar
- *newshape* la forma que sera *int* o una tupla de *int.*

Se puede utilizar de 2 formas:

- Invocandolo como metodo de la libreria

> ***reshape()***

- Invocandolo como "metodo propio" (**metodo asociado a un objeto**) en el Array

> ***ndarray.reshape***


#### Funcion *flatten()*

Crea una copia de un Array, con todos sus elementos a un Array de 1 dimension (vector)

> Por tanto generar Arrays de 1 dimension (vectores) a raiz de Arrays de mas dimensiones (matrices)

#### *Mascaras Booleanas* con los Arrays

Existe otra funcionalidad con los Arrays que nos va a permitir realizar validaciones muy basicas.

Son las llamadas *mascaras booleanas*

Se obtienen cuando en un Array entero, con cualquier forma, **compuesto por valores**, lo comparamos con una expresion **simple de evaluacion**

> ***variable = np.array (["valores enteros"])***
>
> ***variable "=" / "!=" "valor entero buscado***
>
> > Nos devolvera un Array **con la misma forma** (vector, 2 dimensiones,...) en el que indicara **que valor cumple (*True*) o no cumple (*False*) la evaluacion**
