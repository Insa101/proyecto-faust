# Apuntes Pandas 2: Lectura y conversion con ficheros tabulados

La libreria pandas incorpora modulos que permiten trabajar de manera sencilla con ficheros de texto con datos tabulados, aparte de otros tipos.

Los principales ficheros con los que vamos a trabajar son los **archivos .csv**, ya que son ficheros tabulados que **separan los datos con delimitadores**. 

Para poder trabajar con los ficheros, necesitamos en primer lugar **acceder a ellos** convirtiendolos a lenguaje de Python y posteriormente **habra que volver a crear un fichero con su formato** para poder tener los datos fuera de Python.

Las siguientes funciones son para **el acceso a ficheros** y segun los delimitadores haremos uso de una u otra, como aparece en esta tabla:



> En las funciones ***read_csv y read_table*** podemos definir manualmente el caracter delimitador
>

```python
import pandas as pd

dataset_vuelos = pd.read_csv("C:/proyecto-faust/M02_Python_Librerias_Big_Data/UX_Material_Extra_Librerias/U10_datasets/NYC_flights_2013_MINI.csv" , sep = ";")
dataset_vuelos.shape
```

```python
import pandas as pd

dataset_vuelos = pd.read_table("C:/proyecto-faust/M02_Python_Librerias_Big_Data/UX_Material_Extra_Librerias/U10_datasets/NYC_flights_2013_MINI.csv" , sep = ";")
dataset_vuelos.describe()
```

Estas funciones permiten especificar distintos argumentos para poder visualizar especificamente ese contenido.

Estos argumentos son aquellos que se incluyen dentro de la expresion:

> ***"nombre variable" = pd.read_csv/read_table ( "ruta fichero" , "argumento1", "argumento2",...)***

En la siguiente tabla aparecen algunos de los argumentos que se pueden definir principalmente:



```python
import pandas as pd

dataset_vuelos = pd.read_table("C:/proyecto-faust/M02_Python_Librerias_Big_Data/UX_Material_Extra_Librerias/U10_datasets/NYC_flights_2013_MINI.csv" , sep = ";", na_values = "116")
print (" Indico que el valor 116 se reconozca como na_values y el resultado es:")
print(40*"-")
print(dataset_vuelos.loc[0:10,["distance","origin"]])
```

Como mencionabamos antes, es necesario una vez se ha accedido a los ficheros, devolverlos a un **formato de salida fuera de python** en este caso .csv o tabulado.

Para ello utilizaremos ***.to_csv()***

> ***"nombre dataframe.to_csv("nombre fichero", "argumento 1", "argumento 2",...)***
>

Es recomendable utilizar argumentos para exportar **los datos mas relevantes**, aqui aparece unba lista de argumentos utilizable:



```python
import pandas as pd

dataset_vuelos = pd.read_table("C:/proyecto-faust/M02_Python_Librerias_Big_Data/UX_Material_Extra_Librerias/U10_datasets/NYC_flights_2013_MINI.csv" , sep = ";", na_values = "116")
print (" Indico que el valor 116 se reconozca como na_values y el resultado es:")
print(40*"-")
small_data = dataset_vuelos.loc[0:10,["distance","origin"]]
small_data.to_csv("C:/proyecto-faust/M02_Python_Librerias_Big_Data/U2_Libreria_Pandas/prueba_exportacion_csv_dataset_small.csv", sep =";", header =True, index =True, index_label = True)
```
